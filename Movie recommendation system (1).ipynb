```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('movies_dataset.csv')
```

```python
movies
```

```python
movies = movies[['IMDB ID','Title','Genres','Summary','Writers','Director','Cast']]
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].Genres
```

```python
movies.iloc[0].Summary
```

```python
movies['Summary'] = movies['Summary'].apply(lambda x:x.split())
movies['Writers'] = movies['Writers'].apply(lambda x:x.split())
movies['Director'] = movies['Director'].apply(lambda x:x.split())
movies['Cast'] = movies['Cast'].apply(lambda x:x.split())
movies['Genres'] = movies['Genres'].apply(lambda x:x.split())
```

```python
movies
```

```python
movies['Genres'] = movies['Genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['Summary'] = movies['Summary'].apply(lambda x:[i.replace(" ","") for i in x])
movies['Writers'] = movies['Writers'].apply(lambda x:[i.replace(" ","") for i in x])
movies['Director'] = movies['Director'].apply(lambda x:[i.replace(" ","") for i in x])
movies['Cast'] = movies['Cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies
```

```python
movies['Tags'] = movies['Genres'] + movies['Summary'] + movies['Writers'] + movies['Director'] + movies['Cast']
```

```python
movies
```

```python
new_df = movies[['IMDB ID','Title','Tags']]
```

```python
new_df
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x: " ".join(x))
```

```python
new_df
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['Tags'] = new_df['Tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=4000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['Tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['Title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].Title)
```

```python
recommend('And Then I Go')
```

```python
recommend('New York Doll')
```

```python
recommend("Mickey's House of Villains")
```
